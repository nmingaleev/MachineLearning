### Dataset description

The dataset shows prediction of residuary resistance of sailing yachts at the initial design stage. The dataset comprises 308 experiments perfomed at the Delft Ship Hydromechanics Laboratory.
Link: http://archive.ics.uci.edu/ml/datasets/Yacht+Hydrodynamics

```python
import pandas as pd
import numpy as np
```

```python
cols = ['longitude', 'prismatic_coefficient', 'length_displacement_ratio', 'beam_draught_ratio', 'length_beam_ratio', 'froude_number', 'resistance']
df = pd.read_fwf('yacht_hydrodynamics.data',
                  names=cols
                 )
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.set(style='whitegrid', context='notebook')
sns.pairplot(df[cols])
plt.show()
```

From the above plots we can see that there is a quadratic connection between resistance and froude number

```python
cm = np.corrcoef(df[cols].values.T)
plt.figure(figsize = (15,15))
hm = sns.heatmap(cm,
                 cbar=True,
                 annot= True,
                 square=True,
                 annot_kws={'size': 15},
                 yticklabels=cols,xticklabels=cols
                )
plt.show()
```

The heatmap also affirms our assertion.

### Linear regression

```python
from sklearn.model_selection import train_test_split

X = df.iloc[:, :-1].values
y = df['resistance'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

slr = LinearRegression()
slr.fit(X_train, y_train)

y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)
```

```python
def plt_predictions(y_train_pred, y_train, y_test_pred, y_test):
    plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label='Train data')
    plt.scatter(y_test_pred, y_test_pred - y_test, c='lightgreen', marker='s', label='Test data')

    plt.xlabel('Predicted values')
    plt.ylabel('Remains')

    plt.legend(loc='upper left')
    plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='red')
    plt.xlim([-10, 50])
    plt.show()
    
plt_predictions(y_train_pred, y_train, y_test_pred, y_test)
```

```python
from sklearn.metrics import mean_squared_error

def print_mse(y_train_pred, y_train, y_test_pred, y_test):
    MSE_train = mean_squared_error(y_train, y_train_pred)
    MSE_test = mean_squared_error(y_test, y_test_pred)

    print('MSE train: %.3f' % MSE_train)
    print('MSE test: %.3f' % MSE_test)
    
print_mse(y_train_pred, y_train, y_test_pred, y_test)
```

```python
from sklearn.metrics import r2_score

def print_r2(y_train_pred, y_train, y_test_pred, y_test):
    R2_train = r2_score(y_train, y_train_pred)
    R2_test = r2_score(y_test, y_test_pred)

    print('R2 train: %.3f' % R2_train)
    print('R2 test: %.3f' % R2_test)
    
print_r2(y_train_pred, y_train, y_test_pred, y_test)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.35)
ridge.fit(X_train, y_train)

y_train_pred = ridge.predict(X_train)
y_test_pred = ridge.predict(X_test)

plt_predictions(y_train_pred, y_train, y_test_pred, y_test)
print_mse(y_train_pred, y_train, y_test_pred, y_test)
print_r2(y_train_pred, y_train, y_test_pred, y_test)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.05)
lasso.fit(X_train, y_train)

y_train_pred = lasso.predict(X_train)
y_test_pred = lasso.predict(X_test)

plt_predictions(y_train_pred, y_train, y_test_pred, y_test)
print_mse(y_train_pred, y_train, y_test_pred, y_test)
print_r2(y_train_pred, y_train, y_test_pred, y_test)
```

```python
from sklearn.linear_model import ElasticNet

elastic = ElasticNet(alpha=0.2, l1_ratio=1.0)
elastic.fit(X_train, y_train)

y_train_pred = elastic.predict(X_train)
y_test_pred = elastic.predict(X_test)

plt_predictions(y_train_pred, y_train, y_test_pred, y_test)
print_mse(y_train_pred, y_train, y_test_pred, y_test)
print_r2(y_train_pred, y_train, y_test_pred, y_test)
```

Having several models trained we can see that the best result was demonstrated by ElasticNet with R2 = 0.583 on the test data against
Lasso with R2 = 0.564,
Ridge with R2 = 0.554 and
Linear with R2 = 0.549.
Nevertheless 0.583 is a pretty weak result.

```python
#quadratic
regr = regr.fit(X_quad, y)
y_quad_fit = regr.predict(quadratic.fit_transform(X_fit))

quadratic_mse = mean_squared_error(y, regr.predict(X_quad))
quadratic_r2 = r2_score(y, regr.predict(X_quad))

print('MSE = %.2f' % quadratic_mse)
print('R2 = %.2f' % quadratic_r2)
```

As we can see the best result is shown by cubic regression with R2 = 0.98 against quadratic regression with R2 = 0.93

### Random forest regression

```python
X = df.iloc[:, :-1].values
y = df['resistance'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(n_estimators=1000,
                              criterion='mse',
                              random_state=0,
                              n_jobs=-1)

forest.fit(X_train, y_train)

y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)

print_mse(y_train_pred, y_train, y_test_pred, y_test)
print_r2(y_train_pred, y_train, y_test_pred, y_test)
```

```python
plt.scatter(y_train_pred,
           y_train_pred-y_train,
           c='black',
           marker='o',
           s=35,
           alpha=0.5,
           label='Train values'
           )

plt.scatter(y_test_pred,
           y_test_pred - y_test,
           c='lightgreen',
           marker='s',
           s=35,
           alpha=0.7,
           label='Test values'
           )

plt.xlabel('Predicted values')
plt.ylabel('Remains')

plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=65, lw=2, color='red')
plt.xlim([-10, 65])

plt.show()
```

```python
print('Feature importance:')

print('Longitude: %.2f' % forest.feature_importances_[0])
print('Prismatic coefficient: %.2f' % forest.feature_importances_[1])
print('Length displacement ratio: %.2f' % forest.feature_importances_[2])
print('Beam draught ratio: %.2f' % forest.feature_importances_[3])
print('Length beam ratio: %.2f' % forest.feature_importances_[4])
print('Fraude number: %.2f' % forest.feature_importances_[5])
```

### Conclusion

We have tried different regression algorythms such as linear regression (Linear, Ridge, Lasso, ElasticNet), polynomial regression (quadratic, cubic) and random forest regression. The best result was achived by random forest regressor with R2 = 0.998 on the test data. The most important feature according to the random forest is Fraude number with 99 % of importance.
