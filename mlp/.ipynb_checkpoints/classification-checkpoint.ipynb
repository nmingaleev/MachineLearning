У нас есть все данные и мы готовы построить некоторую модель, которая по заданным признакам, описывающим состояние пациента, сможет определить, обнаружен ли у пациента сахарный диабет.  
Чтобы обучить модель, нам необходимо:
 - построить некоторый алгоритм машинного обучения;
 - настроить гиперпараметры этого алгоритма, используя некоторые показатели качества обучения;
 - используя найденные оптимальные гиперпараметры, обучить алгоритм на всех имеющихся данных и получить готовую модель.  

Имея готовую модель, можно использовать её для решения реальных проблем предметной области - ставить диагноз - определять, будет ли обнаружен сахарный диабет у тех или иных пациентов.

Воспользуемся методом отложенной выборки и создадим обучающую и валидационную выборки из подготовленных для обучения данных. При этом сделаем наши выборки (обучающую и валидационную) стратифицированными. Т.е. процентное соотношение объектов различных классов в них должно остаться тем же, каким было в первоначальной выборке.

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, stratify=y, test_size=0.2, shuffle=True, random_state=9)
```

Преобразуем данные таким образом, как это необходимо делать в большинстве случаев для обучения нейронных сетей, когда данные являются разнородными - различные числовые признаки измеряются в различных единицах. Это необходимо для успешного обучения нейронных сетей на такого рода данных.

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_val = ss.transform(X_val)
```

Данные мы подготовили. Теперь мы готовы к созданию архитектуры простой нейронной сети в качестве модели классификации. Используем Keras для того, чтобы в отдельности описать каждый из слоёв нейронной сети и объединить их в единую структуру.  
 - Input - входной слой сети, в котором указывается параметр shape - размерность входного слоя.
 - Dense - это простой полносвзный слой.
 - Для каждого слоя прописываем функцию активации нейронов слоя: *relu = max(0, x)* выход функции - действительное число от 0 до Inf; *sigmoid = 1 / 1 + exp(-x)*, выход функции - действительное число от 0.0 до 1.0.
 - kernel_initializer - вид случайного распределения, с помошью которого инициируются значения весов нейронов слоя. Рекомендуется нормальное распределение.  

 - Функция ошибки. Она выбирается исходя из решаемой задачи и архитектуры сети. Функция выбирается исходя из активаицонной функции выходного слоя сети. В Keras одной из подходящих функций для задачи классификации является функция binary_crossentropy. Фунция кросс энтропии является одной из простых замен функции квадратичной ошибки. Если а - это выход нейрона, а y - это верный ответ, то функция ошибки кросс энтропии выглядит следующим образом *C = -1(sum(y ln(a) + (1 - y) ln(1 - a))) / n*, где n - количество образцов в обучающей выборке. Чем меньше ошибка, тем ближе значение функции кросс энтропии к нулю - чем ближе к нулю, тем лучше.
 - Алгоритм обучения. Основным алгоритмом обучения является алгоритм стохастического градиентного спуска. Но также существет множество его модификаций, ускоряющих сходимость и позволяющих выбираться из локальных минимумов, следовательно, добиваться либо тех же результатов за более короткий цикл обучения, либо, воообще, добиваться более лучших результатов. Сейчас мы используем простейший алгоритм стохастического градиентного спуска.

```python
from keras.models import Model
from keras.layers import Input
from keras.layers import Dense
from keras.layers import Dropout
from keras.optimizers import SGD, Adam

# Baseline
def get_nn():
    input_layer = Input(shape=(8,))
    dense1 = Dense(units=16, activation='relu', kernel_initializer='random_uniform')(input_layer)
    dense2 = Dense(units=16, activation='relu', kernel_initializer='random_uniform')(dense1)
    output_layer = Dense(units=1, activation='sigmoid')(dense2)
    
    model = Model(inputs=input_layer, outputs=output_layer)
    model.compile(loss='binary_crossentropy', optimizer=SGD())
    
    return model


# Modification 1
# def get_nn():
#     input_layer = Input(shape=(8,))
#     dense1 = Dense(units=20, activation='relu', kernel_initializer='random_uniform')(input_layer)
#     dense2 = Dense(units=20, activation='relu', kernel_initializer='random_uniform')(dense1)
#     output_layer = Dense(units=1, activation='sigmoid')(dense2)
    
#     model = Model(inputs=input_layer, outputs=output_layer)
#     model.compile(loss='binary_crossentropy', optimizer=SGD())
    
#     return model


# Modification 2
# def get_nn():
#     input_layer = Input(shape=(8,))
#     dense1 = Dense(units=20, activation='relu', kernel_initializer='random_uniform')(input_layer)
#     dense2 = Dense(units=20, activation='relu', kernel_initializer='random_uniform')(dense1)
#     output_layer = Dense(units=1, activation='sigmoid')(dense2)
    
#     model = Model(inputs=input_layer, outputs=output_layer)
#     model.compile(loss='binary_crossentropy', optimizer=Adam(lr=0.00005))
    
#     return model


# Modification 3
# def get_nn():
#     input_layer = Input(shape=(8,))
#     dense1 = Dense(units=20, activation='relu', kernel_initializer='random_uniform')(input_layer)
#     dropout1 = Dropout(0.5)(dense1)
#     dense2 = Dense(units=20, activation='relu', kernel_initializer='random_uniform')(dropout1)
#     dropout2 = Dropout(0.5)(dense2)
#     output_layer = Dense(units=1, activation='sigmoid')(dropout2)
    
#     model = Model(inputs=input_layer, outputs=output_layer)
#     model.compile(loss='binary_crossentropy', optimizer=Adam(lr=0.00005))
    
#     return model
```

Теперь обучим нашу нейронную сеть!
 - Мы используем обучение с учителем - подадим на вход обучающие матрицу объекты-признаки и вектор меток классов.
 - Пусть количество эпох обучения будет равным 600.
 - Параметр batch_size. Это особый параметр значительно влияющий на процесс обучения. Относительно размера batch'а есть 2 типа обучения: когда для каждого объекта производится прямой проход, рассчёт ошибки и обратный проход; когда для n объектов производится прямой проход, рассчёт и накопление ошибок, затем обратный проход после накопление и усреднения n ошибок. Обучение на batch'ах более эффективно и используется сейчас везде: оно быстрее, оно не позволяет отдельным образцам выбросить процесс обучения в локальные минимумы и изменение функции ошибки проходит более плавно. Выберем значение этого параметра 32.  
 - Validation data - это как раз реализация подхода обучения по отложенной выборке - модель на протяжении эпохи обучается на данных train, после окончания эпохи, она тестируется на данных val. Это нужно для того, чтобы контролировать качество на данных, на которых модель не обучается и обнаруживать переобучение.
 - shuffle - случайное перемешивание выборки после каждой эпохи.

```python
epochs_num = 600

nn = get_nn()
nn.fit(X_train, y_train, epochs=epochs_num, batch_size=32, validation_data=(X_val, y_val), shuffle=True)
```

Данные вывода Keras недостаточно наглядны - здесь сложно понять, как изменяются значения функции ошибки (сложно отловить переобучение; сложно определить, насколько хорошо обучилась модель). Для этого нам необходимо собирать статистику после каждой эпохи обучения. Для этого в Keras существуют специальные объекты Callback'и. В общем смысле Callback - это функция, которая передается в качесте аргумента в другую функцию и затем вызывается в процессе выполнения этой функции. В Keras Callback - это класс, который имеет методы, вызываемые после определённых событий при обучении. В Keras существуют встроенные Callback'и. Используем свой простейший callback для того, чтобы собирать статистику обучения.

```python
from keras.callbacks import Callback

class CustomCallback(Callback):
    def on_train_begin(self, logs={}):
        self.train_loss = []
        self.val_loss = []
        
    def on_epoch_end(self, epoch, logs={}):
        self.train_loss.append(logs.get('loss'))
        self.val_loss.append(logs.get('val_loss'))
```

Теперь, чтобы собрать данные об обучении нейронной сети и записать их в свойства объекта нашего класса CustomCallback, нам необходимо заново обучить нашу модель, передав функции обучения в качестве аргумента наш объект-Callback. Вся статистика во время обучения теперь будет собираться и записываться в переменные-списки, поэтому можно убрать вывод функции fit.  
Есть важный момент - мы не можем обучить ту же модель, которую обучали в первый раз - её параметры (веса нейронов) уже настроены и, вызвав обучение этой модели ещё раз, мы продолжим её обучать. Нам же необходимо обучить её ещё раз с нуля, т.е. нужно заново случайным образом инициализировать веса нейронов - поэтому код создания модели представляет собой функцию, и каждый вызов функции возвращает нам модель со случайно инициализированными весами.

```python
custom_callback = CustomCallback()
nn = get_nn()

nn.fit(X_train, y_train, epochs=epochs_num, batch_size=32, validation_data=(X_val, y_val), 
       shuffle=True, callbacks=[custom_callback], verbose=0)
```

Исопльзуем простую функцию, которая принимает три аргумента: количество эпох обучения, вектор функции ошибки, полученный на обучающей выборке и вектор значений функции ошибки, полученный на валидационной выборке.

```python
import matplotlib.pyplot as plt

%matplotlib inline

def loss_graph(epochs, train_loss, val_loss):
    x_axis = np.linspace(1, epochs, num=epochs)
    
    plt.figure(figsize=(10, 7))
    plt.plot(x_axis, train_loss, color='b', lw=3, alpha=0.7, label='Train Loss')
    plt.plot(x_axis, val_loss, color='r', lw=3, alpha=0.7, label='Val Loss')
    plt.title('Loss graph')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(loc='upper right')
    plt.grid(True)
```

```python
loss_graph(epochs_num, custom_callback.train_loss, custom_callback.val_loss)
```

Давайте теперь попробуем разбить нашу выборку на обучающую и тестовую по-другому - используем другое значение random_state в функции train_test_split. Затем снова обучим модель и построим график функции ошибок.

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, stratify=y, test_size=0.2, shuffle=True, random_state=90)
X_train = ss.fit_transform(X_train)
X_val = ss.transform(X_val)

nn = get_nn()
nn.fit(X_train, y_train, epochs=epochs_num, batch_size=32, validation_data=(X_val, y_val), 
       shuffle=True, callbacks=[custom_callback], verbose=0)

loss_graph(epochs_num, custom_callback.train_loss, custom_callback.val_loss)
```

Попробуем избежать недостатков подхода отложенной выборки и реализуем метод кросс-валидации.

```python
from sklearn.model_selection import StratifiedKFold

folds_num = 5
kf = StratifiedKFold(n_splits=folds_num, shuffle=True, random_state=7)
ss = StandardScaler()

X_trains, X_vals, y_trains, y_vals = [], [], [], []
for train_idx, val_idx in kf.split(X, y):
    X_trains.append(ss.fit_transform(X[train_idx]))
    X_vals.append(ss.transform(X[val_idx]))
    y_trains.append(y[train_idx])
    y_vals.append(y[val_idx])
```

Проведём обучение на кросс-валидации, т.е. получим средее значение функции ошибки на обучении и валидации. Это позволит более точно определить наиболее подходящие значения гиперпараметров алгоритма, при которых достигаются наилучшие средние значения.  
В данном эксперименте таким оптимизируемым гиперпараметром является количество эпох обучения - солько эпох обучения достаточно для того, чтобы модель достигла минимальной ошибки и, при этом, не начала переобучаться.

```python
train_loss = np.zeros((folds_num, epochs_num))
val_loss = np.zeros((folds_num, epochs_num))

for i in range(len(X_trains)):
    nn = get_nn()
    nn.fit(X_trains[i], y_trains[i], epochs=epochs_num, batch_size=32, validation_data=(X_vals[i], y_vals[i]), 
           shuffle=True, callbacks=[custom_callback], verbose=0)
    
    train_loss[i] = custom_callback.train_loss
    val_loss[i] = custom_callback.val_loss
    
train_loss_mean = train_loss.mean(axis=0).ravel()
val_loss_mean = val_loss.mean(axis=0).ravel()
```

```python
loss_graph(epochs_num, train_loss_mean, val_loss_mean)
```

Мы реализовали кросс-валидацию и теперь видим усреднённую оценку изменения функции ошибки для нашей модели. Теперь мы сможем определить наиболее оптимальное количество итераций для обучения модели, которое очень в малой степени будет зависеть от разбиения выборки. Вы можете изменить значение параметра random_state в объекте StratifiedKfold - разбиение выборки для кросс-валидации изменится. Проведя эксперимент на новом разбиении вы увидите, что изменения в функции потерь очень небольшие, что позволяет считать оценку ошибки на кросс-валидации достаточно точной.  

Определим минимальное значение ошибки, которого достигает модель на кросс-валидации и количество эпох, необходимое для достижения этого оптимального значения.

```python
min_val_loss = np.min(val_loss_mean)
optimal_epoch_num = np.argmin(val_loss_mean)

min_val_loss, optimal_epoch_num
```

Как мы по полученным нами результатам сможем определить качество полученного решения? Мы понимаем, что на после определённой эпохи мы получаем наиболее оптимальную конфигурацию модели на имеющемся наборе данных. Но мы не можем определить, насколько хорошо наша модель решает поставленную задачу - насколько хорошо она может определить, есть ли у пациента сахарный диабет. В реальных проектах эти показатели обычно выдвигаются в требованиях, и, если наша модель не позволит их достич, несмотря на низкое значение ошибки, то нам нужно либо улучшать эту модель, либо пробовать совершенно другую, либо дополнительно работать над данными.  
Для этого необходимо реализовать рассчёт некоторых метрик качества. Возьмём самую простую и чаще всего используемую метрику - Accuracy.  
Keras позволяет рассчитывать метрики во время обучения модели, так, что изменение значения метрик можно проследить на протяжении всего обучения модели.

```python
# Функция создания модели, рассчитывающей метрику Accuracy во время обучения
def get_nn_modified():
    input_layer = Input(shape=(8,))
    dense1 = Dense(units=16, activation='relu', kernel_initializer='random_uniform')(input_layer)
    dense2 = Dense(units=16, activation='relu', kernel_initializer='random_uniform')(dense1)
    output_layer = Dense(units=1, activation='sigmoid')(dense2)
    
    model = Model(inputs=input_layer, outputs=output_layer)
    model.compile(loss='binary_crossentropy', optimizer=SGD(), metrics=['accuracy'])
    
    return model


# Callback, содирающий информацию об изменении метрики Accuracy
class CustomCallbackModified(Callback):
    def on_train_begin(self, logs={}):
        self.train_loss = []
        self.val_loss = []
        self.train_acc = []
        self.val_acc = []
        
    def on_epoch_end(self, epoch, logs={}):
        self.train_loss.append(logs.get('loss'))
        self.val_loss.append(logs.get('val_loss'))
        self.train_acc.append(logs.get('acc'))
        self.val_acc.append(logs.get('val_acc'))

        
custom_callback_modified = CustomCallbackModified()

# Кросс-валидация, позволяющая рассччтывать средние значения метрики Accuracy
train_loss = np.zeros((folds_num, epochs_num))
val_loss = np.zeros((folds_num, epochs_num))
train_acc = np.zeros((folds_num, epochs_num))
val_acc = np.zeros((folds_num, epochs_num))

for i in range(len(X_trains)):
    nn = get_nn_modified()
    nn.fit(X_trains[i],
           y_trains[i],
           epochs=epochs_num,
           batch_size=32,
           validation_data=(X_vals[i], y_vals[i]),
           shuffle=True,
           callbacks=[custom_callback_modified],
           verbose=0)
    
    train_loss[i] = custom_callback_modified.train_loss
    val_loss[i] = custom_callback_modified.val_loss
    train_acc[i] = custom_callback_modified.train_acc
    val_acc[i] = custom_callback_modified.val_acc
    
train_loss_mean = train_loss.mean(axis=0).ravel()
val_loss_mean = val_loss.mean(axis=0).ravel()
train_acc_mean = train_acc.mean(axis=0).ravel()
val_acc_mean = val_acc.mean(axis=0).ravel()

# Функция для построения графика изменения значений метрики Accuracy
def accuracy_graph(epochs, train_acc, val_acc):
    x_axis = np.linspace(1, epochs, num=epochs)
    
    plt.figure(figsize=(10, 7))
    plt.plot(x_axis, train_acc, color='g', lw=3, alpha=0.7, label='Train Accuracy')
    plt.plot(x_axis, val_acc, color='orange', lw=3, alpha=0.7, label='Val Accuracy')
    plt.title('Accuracy graph')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(loc='lower right')
    plt.grid(True)
    
    
# Непосредственно построение графиков собранной статистики во время кросс-валидации модели
loss_graph(epochs_num, train_loss_mean, val_loss_mean)
accuracy_graph(epochs_num, train_acc_mean, val_acc_mean)
```

```python
print('Optimal number of epochs: {0}\nValidation loss = {1:.3f}\nValidation accuracy = {2:.3f}'
      .format(optimal_epoch_num, val_loss_mean[optimal_epoch_num], val_acc_mean[optimal_epoch_num]))
```

Определив оптимальное количество эпох (цель, которую мы предследовали, проводя обучение на кросс-валидации), мы можем обучить модель на всём наборе данных, используя подобранные оптимальные параметры модели и обучения - в данном случае - это количество эпох обучения.

```python
X_ready = ss.fit_transform(X)

nn = get_nn()
nn.fit(X_ready, y, epochs=optimal_epoch_num, batch_size=32, shuffle=True, verbose=0)
```

Теперь мы можем загрузить данные для тестирования и проверить качество модели на этих данных (промоделируем процесс работы системы на реальных данных). Нужно не забыть всегда применять масштабирование (нормировку) признаков.

```python
diab_test = pd.read_csv('diabetes_test.csv')
X_test = diab_test.drop(['Outcome'], axis='columns').values
y_test = diab_test['Outcome'].values

X_test = ss.transform(X_test)
```

Получаем ответы модели на тестовых данных.  
Т.к. для выходного слоя сети используется сигмоидная функция (вещественные числа от 0 до 1), т.е. ответами модели являются значения вероятности принадлежности классифицируемого объекта классу "Положительный диагноз", то ответы классификатора необходимо преобразовать в целочисленные по определённому значению порога.

```python
y_test_proba = nn.predict_on_batch(X_test)
y_test_pred = np.where(y_test_proba > 0.5, 1, 0)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_test_pred)
```

```python
import itertools
import seaborn as sns

def plot_confusion_matrix_mpl(cm, classes):
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Матрица ошибок')
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True', fontsize=12)
    plt.xlabel('Predicted', fontsize=12)
    plt.tight_layout()
    

def plot_confusion_matrix_sns(cm):
    fig, ax = plt.subplots(1, 1, figsize=(6, 6))
    sns.heatmap(cm, annot=True, ax=ax, cmap='GnBu')
    ax.set_ylabel('True', fontsize=12)
    ax.set_xlabel('Predicted', fontsize=12)
```

```python
# plot_confusion_matrix_mpl(cm, ['0', '1'])
plot_confusion_matrix_sns(cm)
```

Подсчитаем метрики качества классификации, которые показала наша модель на тестовых данных. Подсчитаем метрики Accuracy (доля верных ответов), Precision (точность), Recall (полнота), F1 (f-мера).  
Это можно сделать, просто подставив значения матрицы несоответствий в формулы. Но также можно использовать инструменты для задач машинного обучения.
